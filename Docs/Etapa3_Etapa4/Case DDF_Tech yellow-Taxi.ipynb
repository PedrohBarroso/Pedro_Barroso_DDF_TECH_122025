```python
import os
import pandas as pd

# Detecta o diretório onde o arquivo atual (.py ou .ipynb) está
diretorio_atual = os.getcwd() 

# Se o arquivo estiver dentro de Docs/Etapa2/, precisamos subir 2 níveis para chegar na raiz
raiz_do_projeto = os.path.abspath(os.path.join(diretorio_atual, "..", ".."))

# Agora, para acessar o dataset, você monta o caminho assim:
caminho_dataset = os.path.join(raiz_do_projeto, "dataset", "yellow_trip.csv")

# Carregamento seguro
df = pd.read_csv(caminho_dataset)

df
```

```python
# Garantindo a qualidade dos dados utilizando a biblioteca Soda-Core
from soda.scan import Scan

# Criando a instância Scan e adicionando o dataframe à instância Scan
scan = Scan()
scan.set_data_source_name("data_quality_taxi")
scan.add_pandas_dataframe(dataset_name="yellow_trip", pandas_df=df, data_source_name="data_quality_taxi")

# Adicionando as verificações Soda à base de dados
scan.add_sodacl_yaml_str("""
checks for taxi_data:
  - row_count >= 100000
  - missing_count(total_amount) = 0
  - min(total_amount) > 0
  - schema:
      warn:
        when wrong column type:
          passenger_count: integer
""")

# 4. Executar
scan.execute()
print(scan.get_logs_text())
```

```python
# Preparação do dataset para Common Data Model (CDM)

# Padronização e Limpeza (Casting)
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['passenger_count'] = df['passenger_count'].fillna(0).astype(int)

# Mapeamento para o Common Data Model
cdm_mapping = {
    'tpep_pickup_datetime': 'event_timestamp',
    'total_amount': 'transaction_value',
    'passenger_count': 'quantity_metric',
    'PULocationID': 'origin_id'
}
df_cdm = df.rename(columns=cdm_mapping)

df_cdm
```
